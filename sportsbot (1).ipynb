```python
import whisper
import torch
import numpy as np
from IPython.display import Audio, display
import os
import time
import sounddevice as sd
import soundfile as sf
import tempfile
# from google.colab import files
```

```python
def load_whisper_model(model_size="medium", max_retries=3):
    for attempt in range(max_retries):
        try:
            print(f"Loading Whisper {model_size} model (attempt {attempt+1})...")
            # Clear CUDA cache if available
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

            # Try to load the model
            model = whisper.load_model(model_size)
            print(f"Successfully loaded {model_size} model.")
            return model
        except RuntimeError as e:
            if "SHA256 checksum does not not match" in str(e):
                print(f"Download error detected. Retrying in 3 seconds...")
                time.sleep(3)
                # Try to remove the corrupt file (model path might vary)
                try:
                    cache_dir = os.path.join(os.path.expanduser("~"), ".cache", "whisper")
                    if os.path.exists(cache_dir):
                        import glob
                        corrupt_files = glob.glob(os.path.join(cache_dir, f"*{model_size}*"))
                        for file in corrupt_files:
                            os.remove(file)
                            print(f"Removed potentially corrupt file: {file}")
                except Exception as clean_err:
                    print(f"Error cleaning cache: {clean_err}")
            else:
                print(f"Error loading model: {e}")

            if attempt == max_retries - 1:
                print("Maximum retries reached. Trying smaller model...")
                if model_size != "tiny":
                    fallback_size = "tiny" if model_size in ["medium", "large"] else "tiny"
                    print(f"Falling back to {fallback_size} model...")
                    return load_whisper_model(fallback_size)
                else:
                    raise RuntimeError(f"Failed to load any Whisper model after {max_retries} attempts.")
```

```python
def record_audio(duration=5, sample_rate=16000):
    print(f"ğŸ¤ Recording for {duration} seconds... Speak now!")

    # Record audio
    recording = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1)
    sd.wait()  # Wait until recording is finished

    # Create a temporary file to store the recording
    temp_file = tempfile.NamedTemporaryFile(delete=False, suffix='.wav')
    temp_filename = temp_file.name

    # Save recording to the temporary file
    sf.write(temp_filename, recording, sample_rate)

    print("âœ… Recording finished!")

    # Play back the recording to verify
    display(Audio(temp_filename, autoplay=False))

    return temp_filename
```

```python
def transcribe_recorded_audio(audio_path, model):
    print("ğŸ”„ Transcribing audio...")

    try:
        # Transcribe audio
        result = model.transcribe(audio_path)

        # Print transcription
        print("ğŸ“ Transcription: " + result["text"])

        # Clean up the temporary file
        os.remove(audio_path)

        return result["text"]

    except Exception as e:
        print(f"âŒ Error during transcription: {e}")
        return ""

# Load the Whisper model once at startup
print("Initializing voice recognition system...")
whisper_model = load_whisper_model("medium")
```

```python
# -*- coding: utf-8 -*-
"""
Sports Chatbot using Gemini (google-genai SDK v1.10.0+) with Web Scraping Tool
- USES MODEL: gemini-2.0-flash
- Implements FAISS-based semantic caching for FAST retrieval using Gemini Embeddings.
- Optimizes Selenium scraping with explicit waits, link timeouts.
- Passes caching wrapper function directly to tools.
- Handles commentary requests by searching for text commentary/play-by-play and synthesizing.
"""

# --- Imports ---
# (Keep all imports as before: os, time, json, re, sys, hashlib, numpy, genai, types, StopCandidateException, selenium, bs4, faiss)
import os
import time
import json
import re
import sys
import hashlib
import numpy as np
from dotenv import load_dotenv

try:
    import google.genai as genai; from google.genai import types
    from google.generativeai.types import StopCandidateException
    print(f"Using google-genai SDK version: {genai.__version__}")
except ImportError as e:
    if "google.generativeai" in str(e): print("WARNING: StopCandEx import failed."); StopCandidateException = Exception
    elif "google.genai" in str(e): print("ERROR: 'google-genai' required."); sys.exit(1)
    else: print(f"ERROR: Import error: {e}"); sys.exit(1)
try:
    from selenium import webdriver; from selenium.common.exceptions import WebDriverException, TimeoutException
    from selenium.webdriver.chrome.options import Options; from selenium.webdriver.chrome.service import Service
    from selenium.webdriver.support.ui import WebDriverWait; from selenium.webdriver.support import expected_conditions as EC
    from selenium.webdriver.common.by import By; from bs4 import BeautifulSoup; from urllib.parse import quote_plus
except ImportError as e: print(f"ERROR: Missing lib ({e}). pip install selenium beautifulsoup4 lxml numpy python-dotenv"); sys.exit(1)
try: import faiss; print(f"Using FAISS version: {faiss.__version__}")
except ImportError: print("ERROR: FAISS not found. pip install faiss-cpu"); sys.exit(1)


# --- Configuration ---
# (Keep Configurations as before)
MODEL_NAME = "gemini-2.0-flash"; EMBEDDING_MODEL = "models/embedding-001"; EMBEDDING_DIM = 768
CACHE_DIR = "search_cache_v6_text_commentary"; CACHE_INDEX_FILE = os.path.join(CACHE_DIR, "faiss_index_v6.idx")
CACHE_MAPPING_FILE = os.path.join(CACHE_DIR, "faiss_mapping_v6.json"); CACHE_SIMILARITY_THRESHOLD = 0.96
WEBDRIVER_TIMEOUT = 20; LINK_LOAD_TIMEOUT = 10; CHROMEDRIVER_PATH = None
os.makedirs(CACHE_DIR, exist_ok=True); print(f"Cache dir: {os.path.abspath(CACHE_DIR)}")


# --- Gemini Client ---
print("\nConfiguring Gemini Client...")
print("\nConfiguring Gemini Client...")
try:
    client = genai.Client(api_key=GOOGLE_API_KEY)
    print("âœ… Gemini Client created.")
except Exception as e: print(f"âŒ ERROR: Failed to create Gemini Client: {e}", file=sys.stderr); sys.exit(1)

# --- Embedding Helper ---
# (Keep get_embedding as before)
def get_embedding(text: str, task_type="RETRIEVAL_QUERY") -> np.ndarray | None:
    global client;
    if not text: return None
    try:
        response = client.models.embed_content(model=EMBEDDING_MODEL, contents=[text], config=types.EmbedContentConfig(task_type=task_type))
        if response.embeddings: return np.array(response.embeddings[0].values).astype('float32')
        else: print(f"   âš ï¸ No embeddings returned for '{text[:30]}...'", file=sys.stderr); return None
    except Exception as e: print(f"   âš ï¸ Embed Err: {e}", file=sys.stderr); return None

# --- Caching Utilities ---
# (Keep normalize_query_for_filename, save_faiss_cache, load_faiss_cache as before)
def normalize_query_for_filename(query: str) -> str: return hashlib.sha1(query.encode()).hexdigest()
faiss_index = None; index_id_to_data = {}; next_faiss_id = 0
def save_faiss_cache():
    global faiss_index, index_id_to_data, next_faiss_id
    if faiss_index and next_faiss_id > 0:
        try:
            print(f"   Saving FAISS index ({next_faiss_id})..."); faiss.write_index(faiss_index, CACHE_INDEX_FILE)
            print(f"   Saving FAISS mapping..."); save_data = {"next_id": next_faiss_id, "mapping": index_id_to_data}
            with open(CACHE_MAPPING_FILE, 'w', encoding='utf-8') as f: json.dump(save_data, f, indent=2)
            print("   âœ… FAISS cache saved.")
        except Exception as e: print(f"   âš ï¸ Error saving FAISS cache: {e}", file=sys.stderr)
def load_faiss_cache():
    global faiss_index, index_id_to_data, next_faiss_id
    if os.path.exists(CACHE_INDEX_FILE) and os.path.exists(CACHE_MAPPING_FILE):
        try:
            print(f"   Loading FAISS index..."); faiss_index = faiss.read_index(CACHE_INDEX_FILE)
            print(f"   Loading FAISS mapping...");
            with open(CACHE_MAPPING_FILE, 'r', encoding='utf-8') as f: loaded_data = json.load(f)
            index_id_to_data = {int(k): v for k, v in loaded_data.get("mapping", {}).items()}; next_faiss_id = loaded_data.get("next_id", 0)
            if faiss_index.ntotal != len(index_id_to_data) or faiss_index.ntotal != next_faiss_id: print(f"   âš ï¸ FAISS cache mismatch. Resetting.", file=sys.stderr); raise ValueError("Cache inconsistency")
            else: print(f"   âœ… FAISS cache loaded ({faiss_index.ntotal} items).")
        except Exception as e: print(f"   âš ï¸ Error loading FAISS cache: {e}. Starting fresh.", file=sys.stderr); faiss_index = faiss.IndexFlatIP(EMBEDDING_DIM); index_id_to_data = {}; next_faiss_id = 0
    else: print("   No existing FAISS cache found. Initializing fresh index."); faiss_index = faiss.IndexFlatIP(EMBEDDING_DIM); index_id_to_data = {}; next_faiss_id = 0

# --- Web Scraping Tool Function ---
# (Keep _perform_web_search_and_extract exactly as before - unchanged)
def _perform_web_search_and_extract(query: str) -> dict:
    """Internal function to perform web scraping with link load timeout."""
    print(f"\n[_perform_web_search_and_extract] Executing OPTIMIZED+TIMEOUT LIVE search for: '{query}'")
    scraped_data = {}; driver = None; main_page_scraped = False; links_scraped_count = 0
    # --- [Keep the full scraping logic here] ---
    try:
        options = Options(); options.add_argument("--headless"); options.add_argument("--disable-gpu"); options.add_argument("--no-sandbox"); options.add_argument("--disable-dev-shm-usage")
        options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
        options.page_load_strategy = 'eager'
        print("   Initializing WebDriver (live)...")
        try:
            if CHROMEDRIVER_PATH and os.path.exists(CHROMEDRIVER_PATH): service = Service(executable_path=CHROMEDRIVER_PATH); print(f"   Using specified ChromeDriver: {CHROMEDRIVER_PATH}")
            else: print("   Using ChromeDriver from system PATH."); service = Service()
            driver = webdriver.Chrome(service=service, options=options); driver.set_script_timeout(WEBDRIVER_TIMEOUT); print("   âœ… WebDriver initialized (live).")
        except WebDriverException as e_init: print(f"   âŒ WebDriver Init Err: {e_init}", file=sys.stderr); scraped_data["error"] = f"WD init failed: {e_init}"; return scraped_data
        except Exception as e_init_other: print(f"   âŒ Unexpected WD Init Err: {e_init_other}", file=sys.stderr); scraped_data["error"] = f"Unexpected init error: {e_init_other}"; return scraped_data
        query_encoded = quote_plus(query); url = f"https://www.bing.com/search?q={query_encoded}&form=QBLH"
        print(f"   Navigating (live) to: {url}"); driver.set_page_load_timeout(WEBDRIVER_TIMEOUT)
        try: driver.get(url)
        except TimeoutException: print(f"   âš ï¸ Timeout loading MAIN Bing page ({WEBDRIVER_TIMEOUT}s).", file=sys.stderr)
        wait = WebDriverWait(driver, WEBDRIVER_TIMEOUT); bing_results_locator = (By.ID, "b_results")
        try: print("   Waiting for Bing results..."); wait.until(EC.presence_of_element_located(bing_results_locator)); print("   Bing results loaded.")
        except TimeoutException: print(f"   âš ï¸ Timeout waiting for Bing results ({WEBDRIVER_TIMEOUT}s).", file=sys.stderr)
        print("   Scraping main page (live)...")
        try:
            soup_main = BeautifulSoup(driver.page_source, "lxml")
            for element in soup_main(["script", "style", "noscript", "header", "footer", "nav", "form", "aside", "button", "input", "textarea", ".b_scopebar", "#b_header", "#b_footer", "[role='navigation']", "[role='search']"]): element.decompose()
            main_page_text = soup_main.get_text(separator="\n", strip=True)
            if main_page_text: scraped_data["Bing Search Results Page"] = main_page_text; print("   âœ… Scraped main page."); main_page_scraped = True
            else: print("   âš ï¸ No main page text."); scraped_data["Bing Search Results Page"] = "[No text]"
        except Exception as e_parse_main: print(f"   âŒ Error parsing main page: {e_parse_main}", file=sys.stderr); scraped_data["Bing Search Results Page"] = f"[Parse Error]"; main_page_scraped = False
        print("   Extracting links (live)..."); main_results_links = []; results = soup_main.select("li.b_algo h2 a[href]"); print(f"   Found {len(results)} links.")
        for a_tag in results:
             title = a_tag.get_text(strip=True); link = a_tag.get("href")
             if link and title and link.startswith(('http://', 'https://')) and '#' not in link: main_results_links.append({"text": title, "link": link})
        links_to_visit = main_results_links[:3]; print(f"   Will visit top {len(links_to_visit)} links (live) with {LINK_LOAD_TIMEOUT}s timeout.")
        body_locator = (By.TAG_NAME, "body")
        for i, item in enumerate(links_to_visit):
            link_url = item["link"]; link_title = item['text']; source_key = f"Source [{i+1}]: {link_url} (Title: {link_title})"
            print(f"   ğŸ”— Visiting link {i+1}: {link_url}")
            try:
                driver.set_page_load_timeout(LINK_LOAD_TIMEOUT); print(f"      driver.get ({LINK_LOAD_TIMEOUT}s)...")
                driver.get(link_url); print(f"      Waiting for body...")
                try: wait.until(EC.presence_of_element_located(body_locator)); print(f"      Body loaded.")
                except TimeoutException: print(f"      âš ï¸ Timeout waiting for body ({WEBDRIVER_TIMEOUT}s).", file=sys.stderr)
                print(f"      Scraping '{link_title}'...")
                soup_link = BeautifulSoup(driver.page_source, "lxml")
                for element in soup_link(["script", "style", "noscript", "header", "footer", "nav", "aside", "form", "button", "input", "textarea", "iframe", "[role='banner']", "[role='complementary']", "[role='contentinfo']", "[role='navigation']", "[role='search']", ".sidebar", ".related-posts", ".comments"]): element.decompose()
                page_text = ""; main_content = (soup_link.find("main") or soup_link.find("article") or soup_link.find("div", role="main") or soup_link.find("section", role="main") or soup_link.find("div", id="main") or soup_link.find("div", id="content") or soup_link.find("div", class_="content") or soup_link.find("div", class_="main-content") or soup_link.find("div", class_="post-content") or soup_link.find("div", class_="entry-content"))
                if main_content: page_text = main_content.get_text(separator="\n", strip=True)
                else: page_text = soup_link.body.get_text(separator="\n", strip=True) if soup_link.body else ""
                if page_text: scraped_data[source_key] = page_text; print(f"      âœ… Scraped link {i+1}."); links_scraped_count += 1
                else: print(f"      âš ï¸ No text link {i+1}."); scraped_data[source_key] = "[No text]"
            except TimeoutException: print(f"      âŒ Timeout loading link {i+1} ({LINK_LOAD_TIMEOUT}s).", file=sys.stderr); scraped_data[source_key] = f"[Timeout Error {LINK_LOAD_TIMEOUT}s]" ; continue
            except Exception as e_link: print(f"      âŒ Error processing link {i+1}: {e_link}", file=sys.stderr); scraped_data[source_key] = f"[Processing Error]"; continue
    except Exception as e_main: print(f"âŒ Error during live search: {e_main}", file=sys.stderr); scraped_data["error"] = f"Live search failed: {e_main}"
    finally:
        if driver: print("   Quitting WebDriver (live)..."); driver.quit(); print("   âœ… WebDriver quit (live).")
    print(f"[_perform_web_search_and_extract] Finished search. Main: {main_page_scraped}. Links: {links_scraped_count}.")
    return scraped_data if isinstance(scraped_data, dict) else {"error": "Tool failed."}

# --- FAISS Caching Wrapper Function ---
# (Keep get_search_results_with_cache exactly as before - unchanged)
def get_search_results_with_cache(query: str) -> dict:
    """Checks semantic cache using FAISS for search results. If not found, performs live search and caches."""
    global faiss_index, index_id_to_data, next_faiss_id
    print(f"\n[get_search_results_with_cache] Received query: '{query}'")
    if not query: return {"error": "Empty query received."}
    new_query_vector = get_embedding(query, task_type="RETRIEVAL_QUERY")
    if new_query_vector is None: print("   âš ï¸ No embedding. Skipping cache."); return _perform_web_search_and_extract(query)
    faiss.normalize_L2(new_query_vector.reshape(1, -1))
    k_nearest = 1; best_match_score = -1.0; best_match_filepath = None; best_match_query = None
    if faiss_index and faiss_index.ntotal > 0:
        try:
            print(f"   Searching FAISS index ({faiss_index.ntotal} vectors)...")
            D, I = faiss_index.search(new_query_vector.reshape(1, -1), k_nearest)
            nearest_neighbor_id = I[0][0]; similarity_score = D[0][0]
            print(f"   FAISS NN ID: {nearest_neighbor_id}, Score: {similarity_score:.4f}")
            if similarity_score >= CACHE_SIMILARITY_THRESHOLD:
                if nearest_neighbor_id in index_id_to_data:
                    cached_query_str, cached_filepath = index_id_to_data[nearest_neighbor_id]
                    if os.path.exists(cached_filepath):
                        print(f"   âœ… FAISS Cache HIT! Query '{query}' ~ '{cached_query_str}'")
                        try:
                            with open(cached_filepath, 'r', encoding='utf-8') as f: cached_data = json.load(f)
                            cached_data.pop('_cached_original_query', None); return cached_data
                        except Exception as e: print(f"   âš ï¸ Error reading cache file {cached_filepath}: {e}. Live.", file=sys.stderr)
                    else: print(f"   âš ï¸ Cache file missing: {cached_filepath}. Live.", file=sys.stderr)
                else: print(f"   âš ï¸ FAISS ID {nearest_neighbor_id} not in mapping. Live.", file=sys.stderr)
            else: best_match_score = similarity_score
        except Exception as e_search: print(f"   âš ï¸ Error searching FAISS: {e_search}. Live.", file=sys.stderr)
    print(f"   âš ï¸ FAISS Cache MISS for query: '{query}' (Best score: {best_match_score:.4f}). Live search.")
    live_data = _perform_web_search_and_extract(query) # Call OPTIMIZED version
    if isinstance(live_data, dict) and "error" not in live_data:
        query_hash = normalize_query_for_filename(query); cache_filepath = os.path.join(CACHE_DIR, f"{query_hash}.json")
        try:
            live_data_to_save = live_data.copy(); live_data_to_save['_cached_original_query'] = query
            print(f"   Saving results to {cache_filepath}")
            with open(cache_filepath, 'w', encoding='utf-8') as f: json.dump(live_data_to_save, f, indent=2)
            print(f"   âœ… Data saved.")
            if new_query_vector is not None:
                current_id = next_faiss_id; faiss_index.add(new_query_vector.reshape(1, -1))
                index_id_to_data[current_id] = [query, cache_filepath]; next_faiss_id += 1
                print(f"   Added vector to FAISS (ID: {current_id}). Size: {faiss_index.ntotal}")
                save_faiss_cache()
            else: print("   âš ï¸ Skipping FAISS add due to embedding failure.")
        except Exception as e: print(f"   âš ï¸ Error saving results/updating FAISS: {e}", file=sys.stderr)
    return live_data if isinstance(live_data, dict) else {"error": "Live search failed."}


# --- Define Tool List (Only Web Search Tool) ---
print("Defining Tool List (Web Search with Cache)...")
sports_tools = [get_search_results_with_cache] # Only include the search tool
print(f"âœ… Tools defined: {[func.__name__ for func in sports_tools]}")


# --- Define System Instruction (Handling Commentary via Web Search) ---
system_instruction = """You are a specialized Sports Information Chatbot. Your primary goal is to provide accurate and relevant answers to user questions about sports using your internal knowledge FIRST. You also have a web search tool (`get_search_results_with_cache`) as a backup.

**Decision Process:**
1.  **Analyze the Query:** Understand the user's question, intent, timeframe, and desired output (e.g., fact, score, stats, *text* commentary).
2.  **Attempt Internal Answer First:** Check internal knowledge thoroughly for rules, history (like IPL 2014 winner), players, concepts, even scorecards if known reliably. If confident, provide the answer directly.
3.  **Identify Need for Web Search Tool:** Use `get_search_results_with_cache` tool ONLY if:
    *   **(A) Live/Very Recent Info Needed:** User asks for info (scores, results, *text* commentary/play-by-play) about games **RIGHT NOW** or **last ~3 days**.
    *   **(B) Internal Knowledge Lacking:** You cannot confidently answer a query about specific/obscure stats, records, less common historical facts, or *detailed text commentary/play-by-play* internally.
    *   **(C) Explicit Commentary Request:** User asks for "commentary", "play-by-play", or to "describe" a specific event/moment in detail. Use the tool to find textual descriptions or live text commentary feeds.
4.  **Formulate Search Query (If Tool Used):** Create a concise query.
    *   For facts/scores/stats: Target that info (e.g., "Man City match result yesterday"). Use history for context (e.g., "IPL score yesterday").
    *   **For Commentary Requests:** Target *text commentary* or *play-by-play descriptions* (e.g., "live text commentary IND vs AUS cricket", "play by play description Messi goal final", "detailed description last over 2014 ipl final").
    **IMPORTANT: Directly initiate the tool call process.**
5.  **Synthesize Tool Results:** The tool returns a dictionary of text snippets. Examine ALL text.
    *   **For Factual Queries:** Extract specific info accurately.
    *   **For Commentary Requests:** Find relevant descriptions or play-by-play text. **Synthesize this text into an engaging, commentary-style response.** Rephrase the factual information to sound like a commentator describing the action. Do NOT just copy text.
    *   **If Info Missing:** If the specific detail (or commentary text) isn't found, state that.
6.  **Handle Tool Failure:** If the tool returns an 'error' key, inform the user the search failed.

**Core Principle:** Internal knowledge first. Web search tool for live/recent info, when knowledge is lacking, OR for finding text to synthesize commentary from.
"""


# --- Load Cache Index at Startup ---
load_faiss_cache()

# --- Chatbot Initialization ---
print(f"\nInitializing Chat Session with model '{MODEL_NAME}'...")
try:
    chat = client.chats.create(
        model=MODEL_NAME,
        config=types.GenerateContentConfig(
            system_instruction=system_instruction,
            tools=sports_tools, # List now only contains the search tool
        )
    )
    print("âœ… Chat session initialized.")
except Exception as e: print(f"âŒ ERROR: Failed to initialize chat session: {type(e).__name__} - {e}", file=sys.stderr); sys.exit(1)


# --- Main Chat Loop (Text Input Only) ---
print("\n--- Sports Chatbot Ready ---")
print(f"Using model: {MODEL_NAME}")
print(f"Embedding model: {EMBEDDING_MODEL} (Dim: {EMBEDDING_DIM})")
print(f"Cache directory: {CACHE_DIR}")
print(f"Cache similarity threshold: {CACHE_SIMILARITY_THRESHOLD}")
print("Ask me about sports! Type 'quit' to exit.")

while True:
    try:
        input_method = input("\nğŸ¤/âŒ¨ï¸ Choose input method (text/voice/file) or type 'quit': ").strip().lower()
        if not input_method:
                continue
        if input_method == 'quit':
                print("\nğŸ¤– Assistant: Goodbye! Thanks for using Sports Chatbot!")
                break
        user_query = ""
        if input_method == 'voice':
                # Record audio and transcribe
                audio_path = record_audio()
                user_query = transcribe_recorded_audio(audio_path, whisper_model)
                print(f"\nğŸ‘¤ You said: {user_query}")

        elif input_method == 'file':
          print("\nğŸ“ Please upload your audio file (mp3, wav, m4a, etc.)")
          uploaded = files.upload()
          if uploaded:
            # Get the filename and create a path
            filename = list(uploaded.keys())[0]
            filepath = os.path.join(os.getcwd(), filename)

            # Display audio player
            print("\nğŸ”Š Audio file uploaded:")
            display(Audio(filepath, autoplay=False))

            # Transcribe the uploaded audio
            print("\nğŸ”„ Transcribing audio file...")
            try:
                result = whisper_model.transcribe(filepath)
                user_query = result["text"]
                print(f"\nğŸ‘¤ Your audio said: {user_query}")

                # Clean up the file if needed
                # os.remove(filepath)  # Uncomment if you want to delete after transcription
            except Exception as e:
                print(f"\nâŒ Error transcribing file: {e}")
          else:
            print("\nâš ï¸ No file was uploaded.")
            continue


        else:
                # Text input
                user_query = input("\nğŸ‘¤ You: ").strip()
                if user_query.lower() == 'quit':
                    print("\nğŸ¤– Assistant: Goodbye! Thanks for using Sports Chatbot!")
                    break
        if not user_query:
                print("I didn't catch that. Please try again.")
                continue
        print(f"\nğŸ¤– Assistant: I received your query about sports: '{user_query}'")
    except EOFError: print("\nInput stream closed."); break
    except KeyboardInterrupt: print("\nInterrupted."); break

    if not user_query: continue
    if user_query.lower() == 'quit': print("\nğŸ¤– Assistant: Goodbye!"); break

    try:
        print("\nğŸ¤– Assistant: Thinking...")
        response = chat.send_message(user_query) # Auto function calling handles dispatch

        if response.text: print(f"\nğŸ¤– Assistant: {response.text}")
        elif response.candidates:
            candidate = response.candidates[0]; print(f"DEBUG: Finish Reason: {candidate.finish_reason.name}")
            if candidate.finish_reason.name == "TOOL_CALLS": tool_call_info = candidate.content.parts[0].function_call if candidate.content and candidate.content.parts else "N/A"; print(f"DEBUG: Requested: {tool_call_info}"); print("\nğŸ¤– Assistant: Tool processing issue.")
            elif candidate.finish_reason.name == "STOP" and not response.text: print("DEBUG: Stopped, no text."); print("\nğŸ¤– Assistant: Processed, but no text response.")
            else: print(f"\nğŸ¤– Assistant: (Finished: {candidate.finish_reason.name}, no text)")
        else: print(f"\nğŸ¤– Assistant: Empty response.")

    except StopCandidateException as e:
        finish_reason_name = e.finish_reason.name if hasattr(e, 'finish_reason') and hasattr(e.finish_reason, 'name') else "UNKNOWN"
        print(f"\nâš ï¸ Resp stopped: {finish_reason_name}", file=sys.stderr)
        if finish_reason_name == "SAFETY": print("   Safety.", file=sys.stderr)
        elif finish_reason_name == "MAX_TOKENS": print("   Max tokens.", file=sys.stderr)
    except Exception as e: print(f"\nâŒ Chat Error: {type(e).__name__} - {e}", file=sys.stderr); print("   Please try again.")# -*- coding: utf-8 -*-
"""
Sports Chatbot using Gemini (google-genai SDK v1.10.0+) with Web Scraping Tool
- USES MODEL: gemini-2.0-flash
- Implements FAISS-based semantic caching for FAST retrieval using Gemini Embeddings.
- Optimizes Selenium scraping with explicit waits, link timeouts.
- Passes caching wrapper function directly to tools.
- Handles commentary requests by searching for text commentary/play-by-play and synthesizing.
"""
```
